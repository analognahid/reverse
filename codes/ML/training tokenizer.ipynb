```python
#!/usr/bin/env python3

import sys,os
from elftools.elf.elffile import ELFFile
from elftools.elf.segments import Segment

```

```python
filePath = './../../binaries/gnuit/src/gitfm'
fh = open(filePath, 'rb')
bin_bytearray = bytearray(fh.read())
```

```python
# https://www.capstone-engine.org/lang_python.html



from capstone import *

from capstone.x86 import *


address_inst = {}
with open(filePath, 'rb') as f:
    elf = ELFFile(f)
    dwarfinfo = elf.get_dwarf_info()
    aranges = dwarfinfo.get_aranges()
    print(len(aranges.entries))
    for arange in aranges.entries:
        print(arange)
    for arange in aranges.entries:

        entry = arange.begin_addr
        exit  = arange.begin_addr + arange.length
        ops = bin_bytearray[entry: exit]

        md = Cs(CS_ARCH_X86, CS_MODE_64)
        md.detail = True
        for inst in md.disasm(ops, entry):

            address_inst[hex(inst.address)] = inst



            print( inst.mnemonic+"  "+inst.op_str)
            (regs_read, regs_write) = inst.regs_access()
```

```python


 
addresses = list(address_inst.keys())
addresses.sort()
sorted_address_inst = {i: address_inst[i] for i in addresses}
 
print(len(list(sorted_address_inst)))
```

```python
# https://towardsdatascience.com/how-to-train-a-bert-model-from-scratch-72cfce554fc6

# NSP- > https://towardsdatascience.com/bert-for-next-sentence-prediction-466b67f8226f#:~:text=Next%20sentence%20prediction%20(NSP,longer%2Dterm%20dependencies%20across%20sentences.

l = list(sorted_address_inst.values())
print(l[2])
```

```python
def get_training_corpus():
    dataset = list(sorted_address_inst.values())
    for start_idx in range(0, len(dataset), 1000):
        samples = dataset[start_idx : start_idx + 1000]
        yield [inst.mnemonic+"  "+inst.op_str for inst in samples]

training_corpus = get_training_corpus()
```

```python
from transformers import AutoTokenizer

old_tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
example = '''mov  edi, eax
            call  0x56bd
            test  eax, eax
            jne  0x61da'''

tokens = old_tokenizer.tokenize(example)
tokens
```

```python
tokenizer = old_tokenizer.train_new_from_iterator(training_corpus, 52000)
```

```python
tokens = tokenizer.tokenize(example)
print(tokens)
```
