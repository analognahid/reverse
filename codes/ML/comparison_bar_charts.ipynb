```python
import csv 
from decimal import Decimal
  
with open('./output/O2.csv', 'r') as read_obj: 
  
    # Return a reader object which will 
    # iterate over lines in the given csvfile 
    csv_reader = csv.reader(read_obj) 
  
    # convert string to list 
    labels, state,our = list(csv_reader) 
    state =[float(x) for x in state]
    our =[float(x) for x in our]
    print(labels, state,our ) 
```

```python

import numpy as np 
import matplotlib.pyplot as plt 
   
# state = [115, 215, 250, 200] 
# our = [114, 230, 510, 370] 
# labels = ['2018','2019','2020','2021']
n=len(state)
r = np.arange(n) 
width = 0.37
font_size = 17
fig, ax = plt.subplots(figsize=(20, 5))


# ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

plt.bar(r, state, color = 'tab:blue', 
        width = width,
        label='StateFormer') 
plt.bar(r + width, our, color = 'tab:orange', 
        width = width, 
        label='SeeType') 
plt.tight_layout()

plt.xlabel("Types",fontsize=font_size) 
plt.ylabel("F1 Scores",fontsize=font_size) 
plt.xticks(rotation=65,fontsize=font_size)
plt.yticks(fontsize=font_size)
# plt.title("Co") 
  
# plt.grid(linestyle='--') 
plt.xticks(r + width/2,labels,fontsize=font_size) 
plt.legend(fontsize=font_size) 
plt.savefig('./output/O2_compare.pdf',dpi=200,bbox_inches="tight")
plt.show() 
```
